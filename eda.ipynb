```python
def plot_all(data):
    varlist = data.select_dtypes(include='number').columns
    print("N_num_cols=",len(varlist))
    nbins=1
    for i in varlist:
        plt.figure(i) 
        if i == 'class_label': nbins = 2
        else: nbins = 40
        sns.distplot(data[i], bins=nbins, kde=False)
        plt.savefig("Plots/"+i+".png", transparent=True)
        plt.close(i)
```

```python
plot_all(dat)
```

```python
def plot_dist_type(data,varlist,acttype = act_type):
    for i in varlist:
        print("variable",i)
        plt.figure(i) 
        for j in acttype: 
            print("activity",j)
            #
        
            kdestat = True
            #if data.loc[data.activity==j][i].mean()==0: 
            #    kdestat = False
            x_weight = len(data.loc[data.activity==j][i])/len(data[i])
            x_weights = [x_weight]*len(data.loc[data.activity==j][i])
            print(data.loc[data.activity==j][i].mean(),data.loc[data.activity==j][i].std(),kdestat, x_weight)
            #hist_kws={'weights': x_weights},
            sns.distplot(data.loc[data.activity==j][i], kde=False,label=j)#.set(xlim=(0.1))
            #sns.kdeplot(data.loc[data.activity==j][i])
            #plt.hist(data.loc[data.activity==j][i],10,histtype=u'step',label=i,density=1)
        plt.legend(prop={'size': 12})
        plt.yscale('log') #,hist=True
        plt.savefig("Plots/type/"+i+".png", transparent=True)
            #plt.close(i)
```

```python
def pair_plot(ds,varlist):
    plt.figure("paiplot")
    sns.pairplot(ds,vars=varlist,hue='activity', diag_kind=None)
    #plt.savefig("Plots/pairplot.png", transparent=True)
    plt.close("paiplot")
```

```python
pair_plot(dat,varlist)
```
